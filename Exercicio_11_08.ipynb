```python
# Bibliotecas
import os
import pandas as pd
import plotly.express as px
```

```python
# 2- Percorrer todos os arquivos de venda base (Pasta de vendas)
list_arq_vendas = os.listdir("/content/drive/MyDrive/CursoPython/Vendas")
display(list_arq_vendas)
```

```python
# 3/4- Importar as bases de venda & Tratar/Compilar as bases de dados
list_tabelas = []
for arquivo in list_arq_vendas:
  if "Vendas" in arquivo:
    tabela = pd.read_csv(f"/content/drive/MyDrive/CursoPython/Vendas/{arquivo}")
    list_tabelas.append(tabela)
tabela_vendas = pd.concat(list_tabelas)
display(tabela_vendas)
```

```python
#5- Definir o produto mais vendido
tabela_produtos = tabela_vendas.groupby('Produto').sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
display(tabela_produtos)
```

```python
#6- Definir o produto responsável pelo maior faturamento
tabela_vendas["Faturamento"] = tabela_vendas["Quantidade Vendida"] * tabela_vendas["Preco Unitario"]
tabela_faturamento = tabela_vendas.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_faturamento)
```

```python
#7- Definir a loja/cidade com maior participação no faturamento
tabela_lojas = tabela_vendas.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tabela_lojas)
```

```python
#8- Criar um dashboard como demonstrativo de resultado
grafico =  px.bar(tabela_lojas, x=tabela_lojas.index, y='Faturamento')
grafico.show()
```
